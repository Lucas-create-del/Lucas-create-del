```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

numeroAnosExperienciaX = np.array([
    4,2,8,1,10,6,3,12,24,17,5,9,7,13,18,11,6,15,20,12
    ])

salarioAnualEmpregadoY = np.array([
    84000,72000,216000,15840,90000,42000,33600,180000,
    74160,120000,48000,66000,88200,114000,36000,49200,
    57600,105600,60000,168000
    ])

coeficiente_angular, intercepto = np.polyfit(
    numeroAnosExperienciaX, salarioAnualEmpregadoY, 1
    )

print(f"Coeficiente Angular: {coeficiente_angular}\n")

print(f"Intercepto: {intercepto}\n")

y_pred = coeficiente_angular * numeroAnosExperienciaX + intercepto
print(f"Previs√£o: {y_pred}")

plt.scatter(
    numeroAnosExperienciaX,salarioAnualEmpregadoY,label="Dados originais"
    )

plt.plot(
    numeroAnosExperienciaX,y_pred, color="red", linewidth=2, label = "Linha de Regress√£o"
    )

plt.xlabel("Sal√°rio anual dos empregados (X)")

plt.ylabel("Ano de exeperi√™ncia dos empregados (Y)")

plt.legend()

plt.title("\nRegress√£o Linear Simples com Numpy")

plt.grid(True)

plt.show()


mediaExperiencia = np.mean(numeroAnosExperienciaX)
mediaSalario = np.mean(salarioAnualEmpregadoY)
print(f"\nM√©dia sobre os anos de experi√™ncias dos empregados: {mediaExperiencia}")
print(f"M√©dia sobre os sal√°rios anuais dos empregados: {mediaSalario}")

medianaExperiencia = np.median(numeroAnosExperienciaX)
medianaSalario = np.median(salarioAnualEmpregadoY)
print(f"\nMediana sobre os anos de experi√™ncias dos empregados: {medianaExperiencia}")
print(f"Mediana sobre os sal√°rios anuais dos empregados: {medianaSalario}")

modaExperiencia = stats.mode(numeroAnosExperienciaX)
modaSalario = stats.mode(salarioAnualEmpregadoY)
print(f"\nMode sobre os anos de experi√™ncias dos empregados: {modaExperiencia}")
print(f"Mode sobre os sal√°rios anuais dos empregados: {modaSalario}")

desvioExperiencia = np.std(numeroAnosExperienciaX)
desvioSalario = np.std(salarioAnualEmpregadoY)
print(f"\nDesvio padr√£o sobre os anos de experi√™ncias dos empregados: {desvioExperiencia}")
print(f"Desvio padr√£o sobre os sal√°rios anuais dos empregados: {desvioSalario}")
```
